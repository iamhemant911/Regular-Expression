```python
import re
import nltk
```

```python
wordlist=list()
for w in nltk.corpus.words.words('en'):
    if w.islower():
        wordlist.append(w)
```

```python
len(wordlist)
```

```python
ed_wordlist=list()
for w in wordlist:
    if re.search(r'ed$',w):
        ed_wordlist.append(w)
```

```python
ed_wordlist
```

```python
eight_char_wordlist=list()
for w in wordlist:
    if re.search(r'^..j..t..$',w):
        eight_char_wordlist.append(w)
```

```python
eight_char_wordlist
```

```python
tokens_list=['email','e-mail','e_mail','e/mail','e|mail','mail']
```

```python
valid_email_list=list()
for w in tokens_list:
    if re.search(r'e(-|_)?mail$',w):
        valid_email_list.append(w)
```

```python
valid_email_list
```

```python
four_letter_list=list()
for w in wordlist:
    #if re.search(r'^[ghi][mno][jlk][def]$',w):
    if re.search(r'^(g|h|i)(m|n|o)(j|l|k)(d|e|f)$',w):    
        four_letter_list.append(w)
```

```python
four_letter_list
```

```python
chatwords=sorted(set(w for w in nltk.corpus.nps_chat.words()))
```

```python
chatwords
```

```python
comedy_words=list()
for w in chatwords:
    if re.search(r'^[ha]+$',w):
        comedy_words.append(w)
```

```python
comedy_words
```

```python
silly_words_list=list()
for w in chatwords:
    if re.search(r'^m+i+n+e+$',w):
        silly_words_list.append(w)
```

```python
silly_words_list
```

```python
treebankwords=sorted(set(nltk.corpus.treebank.words()))
```

```python
floating_point_numbers=list()
for w in treebankwords:
    if re.search(r'^[0-9]+\.[0-9]+$',w):
        floating_point_numbers.append(w)
```

```python
floating_point_numbers
```

```python
dollar_sign_list=list()
for w in treebankwords:
    if re.search(r'^[A-Z]+\$$',w):
        dollar_sign_list.append(w)
```

```python
dollar_sign_list
```

```python
four_digit_list = list()
for w in treebankwords:
    if re.search(r'^[0-9]{4}$',w):
        four_digit_list.append(w)
```

```python
four_digit_list
```

```python
four_digit_list = list()
for w in treebankwords:
    if re.search(r'^[1-2]\d\s{1,2}$',w):
        four_digit_list.append(w)
```

```python
four_digit_list
```
